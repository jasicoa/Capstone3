# Business Understanding

## INTRO

Dalam dunia perhotelan saat ini, pembatalan pemesanan kamar hotel sering terjadi dan dapat dipengaruhi oleh berbagai faktor. Faktor-faktor ini bisa berasal dari pihak hotel, seperti keharusan untuk membayar deposit, adanya daftar tunggu saat melakukan pemesanan, ketersediaan tempat parkir, atau ketidakmampuan hotel untuk memenuhi permintaan khusus dari pelanggan. Pembatalan pemesanan ini dapat membawa dampak signifikan bagi hotel, baik dari segi keuntungan maupun kerugian.

Keuntungan dari kemampuan untuk memprediksi pembatalan pemesanan meliputi peningkatan efisiensi operasional, optimisasi pemanfaatan kamar, dan peningkatan kepuasan pelanggan. Dengan memprediksi kemungkinan pembatalan, hotel dapat mengelola sumber daya mereka dengan lebih baik, mengurangi kerugian finansial akibat kamar kosong, dan memberikan layanan yang lebih responsif dan proaktif kepada pelanggan.

Sebaliknya, ketidakmampuan untuk memprediksi pembatalan dapat membawa kerugian serius bagi hotel. Kerugian ini termasuk potensi pendapatan yang hilang dari kamar yang tidak terisi, peningkatan biaya operasional untuk mengelola pemesanan dan pembatalan, serta penurunan tingkat kepuasan pelanggan karena ketidakmampuan hotel untuk memenuhi kebutuhan mereka secara efektif.

Oleh karena itu, dengan memanfaatkan data dan menganalisis faktor-faktor yang berpengaruh terhadap pembatalan pemesanan, hotel dapat mengambil langkah-langkah strategis untuk meminimalkan pembatalan, meningkatkan efisiensi, dan pada akhirnya, meningkatkan profitabilitas dan kepuasan pelanggan.

## Problem Statement

Seringnya pembatalan pemesanan kamar hotel menimbulkan tantangan dalam pengelolaan sumber daya dan penurunan profitabilitas. Faktor-faktor seperti kebijakan deposit, daftar tunggu, ketersediaan tempat parkir, dan permintaan khusus pelanggan berkontribusi terhadap pembatalan ini. Untuk mengurangi kerugian finansial dan meningkatkan kepuasan pelanggan, diperlukan model prediktif yang dapat mengidentifikasi kemungkinan pembatalan berdasarkan faktor-faktor tersebut, sehingga hotel dapat mengambil tindakan proaktif untuk mengoptimalkan operasional dan layanan mereka.

## Goals

Berdasarkan latar belakang yang telah dijelaskan dan pernyataan masalah yang diidentifikasi, tujuan utama dari penelitian ini adalah untuk mengembangkan model prediktif yang efektif dalam mengidentifikasi kemungkinan pembatalan pemesanan kamar hotel. Model ini akan mempertimbangkan berbagai faktor yang berpengaruh, seperti kebijakan deposit, daftar tunggu, ketersediaan tempat parkir, dan permintaan khusus pelanggan. Melalui pencapaian tujuan-tujuan ini, penelitian ini diharapkan dapat memberikan kontribusi signifikan terhadap peningkatan manajemen dan operasi hotel dalam menghadapi tantangan pembatalan pemesanan.

## Metric Evaluation

Target (y):
- 0 : Tidak Cancel
- 1 : Cancel

| Aktual \ Prediksi | Negatif (0/Tidak Cancel) | Positif (1/Cancel) |
| --- | --- | --- |
| Negatif (0/Tidak Cancel) | Aktual tidak cancel, Prediksi tidak cancel (TN)  | Aktual tidak cancel, Prediksi cancel (FP) |
| Positif (1/Cancel) | Aktual cancel, Prediksi tidak cancel (FN) | Aktual tidak cancel, Prediksi tidak cancel (TP) |


False Negative: Kamar kosong padahal ada pelanggan yang bisa menyewa.

False Positive: Kamar direservasi lebih banyak dari kapasitas, merugikan reputasi hotel dan memengaruhi keputusan pelanggan.

### Attribute Information

| Attribute                    | Data Type          | Description                                                                                     |
|------------------------------|--------------------|-------------------------------------------------------------------------------------------------|
| country                      | Text               | Negara dari mana pelanggan berasal.                                                              |
| market_segment               | Text               | Segmen pasar di mana pemesanan berada.                                                           |
| previous_cancellations       | Integer            | Berapa kali pelanggan membatalkan pemesanan sebelumnya sebelum pemesanan yang saat ini.           |

| total_of_special_requests    | Integer            | Jumlah permintaan khusus yang dibuat oleh pelanggan (misalnya, tempat tidur kembar atau kamar di lantai tinggi). |

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import pandas as pd

categorical_cols = ['market_segment',
 'deposit_type',
 'customer_type',
 'reserved_room_type',
 'continent']
numerical_cols = ['total_of_special_requests',
 'booking_changes',
 'days_in_waiting_list',
 'previous_cancellations',
 'required_car_parking_spaces']

# Define the transformer
transformer = ColumnTransformer([
    # Apply Min-Max scaling to numerical columns
    ('num', MinMaxScaler(), numerical_columns),
    
    # Apply one-hot encoding to categorical columns
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
], remainder='passthrough')

# Define models
logreg = LogisticRegression(random_state=42)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=42)
xgb = XGBClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
gboost = GradientBoostingClassifier(random_state=42)

# Define resampling methods
ros = RandomOverSampler(random_state=42)
rus = RandomUnderSampler(random_state=42)
smote = SMOTE(random_state=42)
near = NearMiss()

# Dictionary of models
models = {
    'KNN': knn,
    'Logistic Regression': logreg,
    'Decision Tree': tree,
    'XGBoost': xgb,
    'Random Forest': rf,
    'Gradient Boosting': gboost,
}

# Dictionary of resampling methods
resamples = {
    'none': None,
    'ros': ros,
    'rus': rus,
    'smote': smote,
    'near miss': near,
}

# Dictionary of metrics
metrics = {
    'balanced_accuracy': 'balanced_accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

# Lists to store results
test_accuracy_arr = []
test_precision_arr = []
test_recall_arr = []
test_f1_arr = []
list_model = []
list_resample = []

# Iterate over all models and resampling methods
for name_resample, resample in resamples.items():
    for name_model, model in models.items():
        crossval = StratifiedKFold(n_splits=5)

        if resample is None:
            estimator = Pipeline([
                ('transform', transformer),
                ('model', model),
            ])
        else:
            estimator = ImbPipeline([
                ('transform', transformer),
                ('resample', resample),
                ('model', model),
            ])

        list_model.append(name_model)
        list_resample.append(name_resample)

        # Cross-validation
        scores = cross_validate(estimator, X_train, y_train, cv=crossval, scoring=metrics)
        
        test_accuracy_arr.append(scores['test_balanced_accuracy'].mean() * 100)
        test_precision_arr.append(scores['test_precision'].mean() * 100)
        test_recall_arr.append(scores['test_recall'].mean() * 100)
        test_f1_arr.append(scores['test_f1'].mean() * 100)

# Create a DataFrame for benchmarking results
model_bench = pd.DataFrame({
    'model': list_model,
    'resample': list_resample,
    'accuracy': test_accuracy_arr,
    'precision': test_precision_arr,
    'recall': test_recall_arr,
    'F1': test_f1_arr
})

# Display the benchmarking results
model_bench
```

**Dalam pipeline ini kita Membangun ColumnTransformer yang akan menerapkan MinMaxScaler pada kolom numerik dan OneHotEncoder pada kolom kategorikal. Kolom numerik akan dinormalisasi ke rentang 0-1, sementara kolom kategorikal akan diubah menjadi beberapa kolom biner. remainder='passthrough' berarti kolom yang tidak ditransformasi akan tetap dipertahankan.**

**Untuk menangani ketidakseimbangan data, digunakan beberapa metode resampling seperti RandomOverSampler, RandomUnderSampler, SMOTE, dan NearMiss. Metode ini membantu memperbaiki distribusi kelas dalam data latih sehingga model dapat belajar dari data yang lebih seimbang.**

```python
from imblearn.pipeline import Pipeline  # Use Pipeline from imblearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

# Define the chosen models from Model Benchmarking
logreg = LogisticRegression(random_state=42)
rus = RandomUnderSampler(random_state=42)
ros = RandomOverSampler(random_state=42)

models = [
    ('Logistic Regression', logreg)
]

resamples = [
    ('RandomUnderSampling', rus),
    ('RandomOverSampler', ros)
]

# List to store results
metrics = []
conf_matrix = []

# Iterate over both models and resampling methods
for name_model, model in models:
    for name_resample, resample in resamples:
        pipeline = Pipeline([
            ('transform', transformer),
            ('resampling', resample),
            ('model', model)
        ])

        # Fit the pipeline on the training data
        pipeline.fit(X_train, y_train)
        
        # Predict on the test data
        y_pred = pipeline.predict(X_test)

        # Calculate and store metrics
        metrics.append([
            accuracy_score(y_test, y_pred) * 100,
            precision_score(y_test, y_pred, average='weighted') * 100,
            recall_score(y_test, y_pred, average='weighted') * 100,
            f1_score(y_test, y_pred, average='weighted') * 100
        ])

        # Store confusion matrix
        conf_matrix.append(confusion_matrix(y_test, y_pred))

# Create a DataFrame to display the metrics
index_names = [f'{name_model} with {name_resample}' for name_model, _ in models for name_resample, _ in resamples]
metrics_df = pd.DataFrame(metrics, columns=['Accuracy', 'Precision', 'Recall', 'F1'], index=index_names)
metrics_df
```

 **Berdasarkan hasil tersebut, kedua pendekatan memberikan performa yang hampir serupa dengan sedikit perbedaan pada nilai metrik evaluasi. Hal ini mengindikasikan bahwa kedua teknik resampling efektif dalam menangani ketidakseimbangan data, namun RandomOverSampler menunjukkan kinerja yang sedikit lebih baik secara keseluruhan. Ini dapat membantu dalam membuat keputusan yang lebih tepat**

```python
# Define Hyperparameter space for the 1st Tuning process
hyperparam_space = {
    'model__penalty': ['l2', 'l1', 'elasticnet'],  # None is not a valid penalty, elastic-net instead of elastic-net
    'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'model__solver': ['liblinear', 'saga', 'lbfgs'],
    'model__max_iter': [100, 500, 1000],
}

logreg = LogisticRegression(random_state=42)
rus = RandomUnderSampler(random_state=42)

pipe_model = ImbPipeline([
    ('prep', transformer),
    ('resampling', rus),
    ('model', logreg)
])

# Define Grid Search
gridsearch = GridSearchCV(
    estimator=pipe_model,
    param_grid=hyperparam_space,  # specifies the range of hyperparameter values to be explored in the grid search
    cv=5,  # specifies the number of folds to be used for cross-validation to evaluate the model
    scoring='recall',   # defines the metric used to evaluate the hyperparameter combinations and select the best model
    n_jobs=-1   # To speed up the process by utilizing multiple cores of the machine
)

gridsearch.fit(X_train, y_train)
```

```python
# Output the best parameters and best score
best_params = gridsearch.best_params_
best_score = gridsearch.best_score_

print("Best Parameters:", best_params)
print("Best Recall Score:", best_score)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import pandas as pd

categorical_cols = ['market_segment',
 'deposit_type',
 'customer_type']
numerical_cols = ['total_of_special_requests',
 'previous_cancellations',
 'required_car_parking_spaces']

# Define the transformer
transformer = ColumnTransformer([
    # Apply Min-Max scaling to numerical columns
    ('num', MinMaxScaler(), numerical_columns),
    
    # Apply one-hot encoding to categorical columns
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
], remainder='passthrough')

# Define models
logreg = LogisticRegression(random_state=42)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=42)
xgb = XGBClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
gboost = GradientBoostingClassifier(random_state=42)

# Define resampling methods
ros = RandomOverSampler(random_state=42)
rus = RandomUnderSampler(random_state=42)
smote = SMOTE(random_state=42)
near = NearMiss()

# Dictionary of models
models = {
    'KNN': knn,
    'Logistic Regression': logreg,
    'Decision Tree': tree,
    'XGBoost': xgb,
    'Random Forest': rf,
    'Gradient Boosting': gboost,
}

# Dictionary of resampling methods
resamples = {
    'none': None,
    'ros': ros,
    'rus': rus,
    'smote': smote,
    'near miss': near,
}

# Dictionary of metrics
metrics = {
    'balanced_accuracy': 'balanced_accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

# Lists to store results
test_accuracy_arr = []
test_precision_arr = []
test_recall_arr = []
test_f1_arr = []
list_model = []
list_resample = []

# Iterate over all models and resampling methods
for name_resample, resample in resamples.items():
    for name_model, model in models.items():
        crossval = StratifiedKFold(n_splits=5)

        if resample is None:
            estimator = Pipeline([
                ('transform', transformer),
                ('model', model),
            ])
        else:
            estimator = ImbPipeline([
                ('transform', transformer),
                ('resample', resample),
                ('model', model),
            ])

        list_model.append(name_model)
        list_resample.append(name_resample)

        # Cross-validation
        scores = cross_validate(estimator, X_train, y_train, cv=crossval, scoring=metrics)
        
        test_accuracy_arr.append(scores['test_balanced_accuracy'].mean() * 100)
        test_precision_arr.append(scores['test_precision'].mean() * 100)
        test_recall_arr.append(scores['test_recall'].mean() * 100)
        test_f1_arr.append(scores['test_f1'].mean() * 100)

# Create a DataFrame for benchmarking results
model_bench = pd.DataFrame({
    'model': list_model,
    'resample': list_resample,
    'accuracy': test_accuracy_arr,
    'precision': test_precision_arr,
    'recall': test_recall_arr,
    'F1': test_f1_arr
})

# Display the benchmarking results
model_bench
```

## Threshold Optimization

**Tujuan dari threshold optimization adalah untuk menemukan nilai ambang batas yang terbaik untuk memaksimalkan performa model klasifikasi. Alih-alih menggunakan threshold standar (biasanya 0.5) untuk memutuskan kelas, kita mencoba berbagai nilai ambang untuk menentukan pada titik mana model memberikan kinerja terbaik, diukur dengan metrik seperti F1 score. Dengan demikian, kita dapat menyesuaikan model agar lebih akurat dan relevan sesuai kebutuhan bisnis, memastikan keputusan klasifikasi yang optimal berdasarkan probabilitas prediksi yang dihasilkan oleh model.**

```python
from sklearn.metrics import recall_score

# Menggunakan model terbaik dari Grid Search
model = refined_gridsearch.best_estimator_
model.fit(X_train, y_train)

# Rentang nilai threshold dari 0.01 hingga 0.98
th_value = [i/100 for i in range(1, 99)]
score_list = []

# Iterasi untuk setiap nilai threshold
for i in th_value:
    # Prediksi probabilitas
    y_prob = model.predict_proba(X_test)[:, 1]
    # Mengubah probabilitas menjadi prediksi berdasarkan threshold
    y_pred = np.where(y_prob > i, 1, 0)
    # Menghitung recall
    score_temp = recall_score(y_test, y_pred)
    # Menyimpan nilai recall
    score_list.append(score_temp)
```
